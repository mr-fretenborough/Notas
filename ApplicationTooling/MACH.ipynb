# MACH

This acronym stands for a collection of development concepts 

The acronym MACH stands for:
1. [Microservices](#microservices)
2. [APIs]()
3. [Cloud-Native]()
4. [Headless]()

## [Microservices]()

Microservice are small business services that can work together and can be deployed autonomously/independently. These services communicate with each other by talking over the network and bring many advantages with them. One of the biggest advantages is that they can be deployed independently. However, it offers the opportunity to work with many different technologies.

Furthermore, these services are built around business capabilities and independently deployable by fully automated deployment process. There is a minimum of centralized management of these services, which may be written in different programming languages and use different data storage technologies.

Generally, MSs have their own tech stack, communicate over a combination of REST APIs, event steams, and message brokers, and are organized by business capability (often referred to as bounded context).

### Characteristics

Microservices should be small, independent, and loosely coupled such that each service is a seperate codebase that is deployed and maintained by its own team--requiring little interaction with other teams.

These MS are also responsible for persisting their own data or external state. 

They also communicate with each other using well-defined APIs. However, services don't need to know what another service does with some information, only that it is needed. This also allows each service to operate with a different tech stack, set of libraries, or frameworks.

## [Application Programming Interfaces]()

API technology connects modules to create an infinitely-flexible ecosystem that can be re-constructed to be completely customer-centric, even as trends and technologies shift.

## [Cloud-Native Software]()

Cloud-native software is commonly delivered as a service by a third-party vendor, so it is always up to date and ready to scale with less cost and time than on-site software.

## [Headless Architecture]()

Headless architecture decouples front-end presentation and back-end logic, streamlining interoperability and enabling channel-independent and framework-agnostic content operations.
